```python
# import libraries packages
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics import r2_score
```

```python
# reading csv file and understanding data outline
df = pd.read_csv("day.csv")
df.head()
df.info()
```

```python
# dropping unnecesssary columnsregistered and casual
del df['casual']
del df['registered']
```

```python
# informatoin after dropping 'casual' and 'registered' columns
df.info()
```

```python
# checking null values
df.isnull().sum()
```

```python
# mapping catagorical variaable 
df['season'] = df['season'].map({1:'spring', 2:'summer', 3:'fall', 4:'winter'})
df['mnth'] = df['mnth'].map({1:'jan', 2:'feb', 3:'mar', 4:'apr',5:'may', 6:'jun', 7:'jul', 8:'aug',9:'sep', 10:'oct', 11:'nov', 12:'dec'})
df['weekday'] = df['weekday'].map({1:'sun', 2:'mon', 3:'tue', 4:'wen',5:'thu', 6:'fri', 7:'sat'})
df['weathersit'] = df['weathersit'].map({1: 'Clear', 2: 'Mist + Cloudy', 3: 'Light Snow', 4: 'Snow + Fog'})
```

```python
df.head()
```

```python
# EDA : pair plotting numeric variables
vars = ['temp', 'atemp', 'hum', 'windspeed', 'cnt']
sns.pairplot(df[vars])
plt.show()
```
